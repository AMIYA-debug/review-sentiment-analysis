```python
from tensorflow.keras.preprocessing.text import one_hot

sent=[
   " ilike to play football",
    " football is a great sport",
    " do you like to play football",
]
voc_size=10000
one_hot_rep=[one_hot(words,voc_size) for words in sent]
one_hot_rep
```

```python
import tensorflow as tf
tokenizer= tf.keras.layers.TextVectorization(
   max_tokens=10000,
   output_mode='int',
   output_sequence_length=6
)

tokenizer.adapt(sent)
output=tokenizer(sent)
output
```

```python
voc_size=len(tokenizer.get_vocabulary())
input_len=len(output[0])
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN
model=Sequential()
model.add(tf.keras.layers.Embedding(input_dim=voc_size,output_dim=16,input_length=input_len) )
model.add(SimpleRNN(16))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
import numpy as np
labels = np.array([1, 0, 1], dtype=np.float32)
output= tf.cast(output, tf.int32)
history=model.fit(output,labels,epochs=10)
```

```python
snet=["football is a good sport"]
snet=tokenizer([snet])
op=model.predict(snet)  
if op>0.5:
      print("positive review")
else:
      print("negative review")     
```
